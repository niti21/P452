```python
#Q1 using cholesky decomposition


with open('a1.txt', 'r') as x1:
    a = [[int(g) for g in line.split(',')] for line in x1]
print(a) 
print()
b1 = open('b1.txt', 'r')
b= b1.read().split(',')
for i in range(0,len(b)):
    b[i]=float(b[i])
print(b)
print()

A = l.cholesky(a)
At = l.transpose(a)
print()

#forward subsitution
B1 = l.forward_sub(A,b)
print('After forward subsitution Matrix is: \n', B1)
print()

#backward subsitution
B2=l.backward_sub(At,B1)
print('After backward subsitution the ANSWER matrix is: \n',B2)
```

```python
#QUESTION 3

with open('a3.txt', 'r') as f1:
    a3 = [[int(num) for num in line.split(',')] for line in f1 ]
A = np.array(a3)

# Given matrix A and vector b
print('A = ', A)
print()

b = np.array([-5/3, 2/3, 3, -4/3, -1/3, 5/3])
print("b = ",b)
print()

# Initial guess for the solution
x0 = [1,1,1,1,1,1]

# Solve using Conjugate Gradient method
x = l.conjugate_gradient(a3, b, x0)
print(f"Solution vector by conjugate gradient method: " )
print(x)
print()


# Calculate the inverse using the solution x

# Create an empty list to store the columns of the inverse matrix
C_inv = []

# Iterate over each column of the identity matrix
for i in range(len(b)):
    # Extract the i-th column of the identity matrix (e_i)
    I_i = np.eye(len(b))[:, i]
    
    # Solve the equation Ax = e_i using Conjugate Gradient method with the solution x
    i_col1 = l.conjugate_gradient(a3, I_i, x0, tol=0.0001)
    
    # Append the solution (column of the inverse matrix) to the list
    C_inv.append(i_col1)

# Stack the columns of the inverse matrix horizontally to form the complete inverse matrix
A_inv = np.column_stack(C_inv)

print("Inverse of A calculated using Conjugate Gradient method:")
print(A_inv)

I =  np.eye(len(A))
check = np.allclose(np.dot(A, A_inv),I)
# Check the correctness of the computed inverse
print("\nCorrectness of the computed inverse:", check )
print()

# Verify the solution
print('Verified solution:')
print(l.gauss_jordan(a3,b))
print()
A_inv_actual = np.linalg.inv(A)
print("\nVerified Inverse of A using numpy:")
print(A_inv_actual)

```

```python
# QUESTION 4

# Define the size of the linear system
N = 50
# Initial guess for the solution
x0 = np.zeros(N)
# Generate the right-hand side vector b (identity matrix)
Ib = np.eye(N)

# Generate the given matrix A
A_ij = l.matrix_A_ij(x0)
print("Input matrix A:")
print((A_ij))
print()

# Compute the inverse of matrix A using Conjugate Gradient method
A_inv = l.conjugate_inv(l.matrix_A_ij, Ib, x0)

# Display the inverse matrix A_inv
print("Inverse of matrix A:")
print((A_inv))
print()

# Plot the inverse matrix
plt.figure(figsize=(8, 6))
plt.imshow(A_inv, cmap='ocean')
#plt.plot(A_ij)
plt.colorbar(label='Element Value')
plt.title('Inverse Matrix')
plt.xlabel('Column Index')
plt.ylabel('Row Index')
plt.show()

# Create a meshgrid for plotting
x = np.arange(N)
y = np.arange(N)
X, Y = np.meshgrid(x, y)

# Plot the 3D surface of the inverse matrix
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, A_inv, cmap='viridis')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
ax.set_title('3D Surface Plot of Inverse Matrix A')
ax.set_xlabel('Column Index')
ax.set_ylabel('Row Index')
ax.set_zlabel('Value')
plt.show()

```
