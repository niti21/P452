```python
import matplotlib.pyplot as plt 
import numpy as np
from scipy import fftpack
from utils import property_crater, parameters
from surface_model import SurfaceModel

```

Lunar topographic map generation:

The aim is to generate topographic map with descriptions about classes and parameters and some example maps. 
The map is given by `SurfaceModel` class with parametric setting by `property_crater` and `parameters` classes.



```python
#Call `property_crater` to set how craters are distributed (*e.g.*, random distribution), geometric characteristics (*e.g.*, normal shape, central-mound shape), and their sizes.


crater_prop = property_crater(
    distribution="random", # types of crater distribution in the map
    geometry="normal", # geometry characteristic
    min_D=1, # minimum crater range
    max_D=5, # maximum crater range
)
```

```python
#Call `parameters` to provide parameters for grid map, such as resolution and number of grids in each axis as well as for fractal terrain modeling, such as roughness exponent.

param = parameters(
    n=200, # number of grid in one axis
    res=1, # grid resolution [m]
    re=0.8, # roughness exponent for fractal surface (0 < re < 1)
    sigma=5, # amplitude gain for fractal surface
    is_fractal=True, # choose to apply fractal surface
    is_crater=True, # choose to apply crater shape
    crater_prop=crater_prop # crater proporty denoted as "property_crater" data structure
)
```

```python
grid_map = SurfaceModel(param=param)
grid_map.set_terrain_env()

grid_map.plot_maps()
plt.show()
plt.savefig('lunar_random.jpg')
```
