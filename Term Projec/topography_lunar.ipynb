```python
import matplotlib.pyplot as plt 
import numpy as np
from scipy import fftpack
import Surface_Model as SM_lib
```

```python
"""

Lunar topographic map generation:

The aim is to generate topographic map with descriptions about classes and parameters and some example maps. 
The map is given by `Surface_Generation` class with parametric setting by `property_crater` and `parameters` classes.

"""
```

```python
"""
Call `property_crater` to set 
-how craters are distributed (*e.g.*, random distribution), 
-geometric characteristics (*e.g.*, normal shape, central-mound shape), and their sizes.

"""

crater_prop = SM_lib.property_crater(
    distribution="random", # types of crater distribution in the map
    geometry="normal", # geometry characteristic
    min_D=1, # minimum crater range
    max_D=5, # maximum crater range
)
```

```python
"""
Call `parameters` to provide parameters for grid map, 
such as resolution and number of grids in each axis as well as for fractal terrain modeling, such as roughness exponent.
"""

param = SM_lib.parameters(
    n=200, # number of grid in one axis
    res=1, # grid resolution [m]
    re=0.8, # roughness exponent for fractal surface (0 < re < 1)
    sigma=5, # amplitude gain for fractal surface
    is_fractal=True, # choose to apply fractal surface
    is_crater=True, # choose to apply crater shape
    crater_prop=crater_prop # crater proporty denoted as "property_crater" data structure
)
```

```python
grid_map = SM_lib.Surface_Generation(param=param)
grid_map.set_terrain()

grid_map.plot_maps()
#plt.savefig('Random crater')
plt.show()
```

```python
"""
1. Single crater visualization - set `distribution="single"`

"""

crater_prop = SM_lib.property_crater(
    distribution="single", # types of crater distribution in the map
    geometry="normal", # geometry characteristic
    min_D=60, # minimum crater range
)

param = SM_lib.parameters(
    n=200, # number of grid in one axis
    res=1, # grid resolution [m]
    re=0.8, # roughness exponent for fractal surface (0 < re < 1)
    sigma=5, # amplitude gain for fractal surface
    is_fractal=True, # choose to apply fractal surface
    is_crater=True, # choose to apply crater shape
    crater_prop=crater_prop # crater proporty denoted as "property_crater" data structure
)

grid_map = SM_lib.Surface_Generation(param=param)
grid_map.set_terrain()

grid_map.plot_maps()
plt.savefig('Single crater')
plt.show()
```

```python
"""
2. No crater visualization - set `is_crater=False`

"""
param = SM_lib.parameters(
    n=200, # number of grid in one axis
    res=1, # grid resolution [m]
    re=0.8, # roughness exponent for fractal surface (0 < re < 1)
    sigma=5, # amplitude gain for fractal surface
    is_fractal=True, # choose to apply fractal surface
    is_crater=False, # choose to apply crater shape
)

grid_map = SM_lib.Surface_Generation(param=param)
grid_map.set_terrain()

grid_map.plot_maps()
plt.savefig('No crater')
plt.show()
```

```python
"""
3. Tune terrain roughness - adjust `re` and `sigma`.

"""

crater_prop = SM_lib.property_crater(
    distribution="random", # types of crater distribution in the map
    geometry="normal", # geometry characteristic
    min_D=1, # minimum crater range
    max_D=5, # maximum crater range
)

param = SM_lib.parameters(
    n=200, # number of grid in one axis
    res=1, # grid resolution [m]
    re=0.8, # roughness exponent for fractal surface (0 < re < 1)
    sigma=10, # amplitude gain for fractal surface
    is_fractal=True, # choose to apply fractal surface
    is_crater=True, # choose to apply crater shape
    crater_prop=crater_prop # crater proporty denoted as "property_crater" data structure
)

grid_map = SM_lib.Surface_Generation(param=param)
grid_map.set_terrain()

grid_map.plot_maps()
plt.savefig('terrain roughness')
plt.show()
```
